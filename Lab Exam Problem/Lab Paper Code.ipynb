```python
from openai import OpenAI
import os                   

# 1. SETUP: Configure Client
# OpenRouter uses the OpenAI SDK. we just change the 'base url'.
client = OpenAI(
    base_url = "https://openrouter.ai/api/v1",
    api_key = "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", # Replace with your actual api key.
)

def get_ai_response(user_query):
    try:
        # 2. SEND INPUT: Create a chat completion
        # We specify the model string exactly as OpenRouter lists it.
        completion = client.chat.completions.create(
            extra_headers={
                "HTTP-Referer": "https://localhost:8000", # Optional: Your Site URL
                "X-Title": "My Simple Python App", # Optional: Your app name
            },
            model="openai/gpt-4o-mini",
            messages=[
                {
                    "role": "user",
                    "content": user_query
                }
            ]
        )

        # 3. HANDLE RESPONSE: Extract the message content
        return completion.choices[0].message.content

    except Exception as e:
        return f"Error: {e}"

def main():
    print("---üåç OpenRouter AI Bot (Type 'quit' to exit)---")

    while True:
        user_input = input("\n Ask Anything: ")

        if user_input.lower() == "quit":
            print("Exiting")
            break

        print("waiting for response...")
        answer = get_ai_response(user_input)

        print(f"\n AI says: \n{answer}")


if __name__ == "__main__": 
    main()
```
