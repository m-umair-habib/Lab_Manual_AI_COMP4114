### 4. Tasks

1. **Ask the user to enter a city name.**  

2. **Call OpenWeather API**  
   - Send a request with the city name to get weather data.  

3. **Extract local time** from the response  
   - Use `dt` and `timezone` fields from the API response.  

4. **Display the information:**  
   - City name  
   - Local time  
   - Current weather  

5. **Repeat the process**  
   - Continue until the user types "bye".


```python
import requests
import datetime
import os
```

```python
api_key = "a74d4d5e3e3ed65877a442a157b****"
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"
```

```python
def get_weather_json(city_name):
    params = {
        "q": city_name,
        "appid": api_key,
        "units": "metric"
    }
    r = requests.get(BASE_URL, params=params)
    return r.json()
```

```python
def local_time(json_data):
    dt_utc = int(json_data["dt"])
    tz_shift = int(json_data["timezone"])
    utc_dt = datetime.datetime.utcfromtimestamp(dt_utc)
    return utc_dt + datetime.timedelta(seconds=tz_shift)
```

```python
def chatbot():
    print("Chatbot: I can show local time & weather. Type a city name or 'bye' to exit.")
    while True:
        city = input("You: ").strip()
        if city.lower() == "bye":
            print("Chatbot: Goodbye!")
            break
        data = get_weather_json(city)
        if data.get("cod") != 200:
            print("Chatbot: Sorry, could not find that city.")
            continue
        local_dt = local_time(data)
        city_name = data['name'] + ", " + data['sys']['country']
        weather = data['weather'][0]['description']
        temp = str(data['main']['temp']) + "°C"
        
        print("Chatbot: Local time in " + city_name + " is " + local_dt.strftime("%Y-%m-%d %H:%M:%S"))
        print("Chatbot: Weather: " + weather + ", Temperature: " + temp)

chatbot()
```
