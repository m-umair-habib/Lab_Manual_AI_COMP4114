### 3. Lab Activities

#### Activity 1 – Reflex Chat Agent

**Task:** Write a Python program that:  

1. Greets the user.  
2. Responds to simple questions such as:  
   - "hello"  
   - "your name"  
   - "how are you"  
3. Ends when the user types "bye".  


```python
# Simple Reflex Chat Agent

print("AI Agent: Hello! I am your simple chat agent. Type 'bye' to exit.")

while True:
    user = input("You: ").lower()
    if user in ["hi", "hello", "hey"]:
        print("AI Agent: Hello! How are you?")
    elif "how are you" in user:
        print("AI Agent: I am just a program, but I am doing great!")
    elif "your name" in user:
        print("AI Agent: I am a simple AI Chat Agent.")
    elif "bye" in user:
        print("AI Agent: Goodbye! Have a great day!")
        break
    else:
        print("AI Agent: Sorry, I don't understand that yet.")
```

#### Activity 2 – Memory-based Chat Agent

**Task:** Modify your program so that the agent can remember the user’s name.

```python
# Chat Agent with Memory (Model-Based)

print("AI Agent: Hello! I am your chat agent. What's your name?")
user_name = input("You: ")

print("AI Agent: Nice to meet you, " + user_name + "! Type 'bye' to exit.")

while True:
    user = input(user_name + ": ").lower()
    
    if "how are you" in user:
        print("AI Agent: I'm fine, " + user_name + "! Thanks for asking.")
    elif "weather" in user:
        print("AI Agent: I can't check live weather yet, but let's assume it's sunny!")
    elif "bye" in user:
        print("AI Agent: Goodbye, " + user_name + "! Take care!")
        break
    else:
        print("AI Agent: Hmm, I don't know how to respond to that, " + user_name + ".")
```
