### 4. Lab Tasks

**Task 1: Create a Simple Rule-Based Expert System (Python)**


```python
# This example identifies an animal based on user input.

print("Animal Expert System")

feathers = input("Does the animal have feathers? (yes/no): ")
flies = input("Can it fly? (yes/no): ")

if feathers == "yes" and flies == "yes":
    print("The animal might be a Bird.")
elif feathers == "yes" and flies == "no":
    print("The animal might be a Penguin.")
else:
    print("The system cannot classify the animal.")
```

### 4. Lab Tasks

**Task 2: Diagnostic Expert System (Simple Medical Example)**


```python
fever = input("Do you have fever? (yes/no): ")
cough = input("Do you have cough? (yes/no): ")

if fever == "yes" and cough == "yes":
    print("Possible Condition: Flu")
elif fever == "yes" and cough == "no":
    print("Possible Condition: Viral Infection")
else:
    print("No major symptoms detected.")
```

### 4. Lab Tasks

**Task 3: Knowledge Base Creation (Rule Writing)**


```python
print("Weather Expert System")

temperature_high = input("Is the temperature high? (yes/no): ")
humidity_high = input("Is the humidity high? (yes/no): ")
temperature_low = input("Is the temperature low? (yes/no): ")
wind_strong = input("Is the wind strong? (yes/no): ")
clouds_dark = input("Are the clouds dark? (yes/no): ")
snowfall = input("Is there snowfall? (yes/no): ")
visibility_low = input("Is visibility low? (yes/no): ")

print("\n--- Weather Prediction ---")

if temperature_high == "yes" and humidity_high == "yes":
    print("Weather Condition: Heatwave")

elif temperature_low == "yes" and wind_strong == "yes":
    print("Weather Condition: Cold Wave")

elif clouds_dark == "yes" and humidity_high == "yes":
    print("Weather Condition: Rain Likely")

elif temperature_low == "yes" and snowfall == "yes":
    print("Weather Condition: Snowstorm")

elif wind_strong == "yes" and visibility_low == "yes":
    print("Weather Condition: Dust Storm")

else:
    print("Weather Condition: Normal or Uncertain")
```
