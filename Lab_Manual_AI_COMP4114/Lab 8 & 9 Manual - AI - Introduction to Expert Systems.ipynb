1. **Task 1: Create a Simple Rule-Based Expert System (Python)**


```python
# This example identifies an animal based on user input.
```

```python
print("Animal Expert System")

feathers = input("Does the animal have feathers? (yes/no): ")
flies = input("Can it fly? (yes/no): ")

if feathers == "yes" and flies == "yes":
    print("The animal might be a Bird.")
elif feathers == "yes" and flies == "no":
    print("The animal might be a Penguin.")
else:
    print("The system cannot classify the animal.")
```

**Task 2: Diagnostic Expert System (Simple Medical Example)**

```python
print("Simple Medical Expert System")

fever = input("Do you have fever? (yes/no): ")
cough = input("Do you have cough? (yes/no): ")

if fever == "yes" and cough == "yes":
    print("Possible Condition: Flu")
elif fever == "yes" and cough == "no":
    print("Possible Condition: Viral Infection")
else:
    print("No major symptoms detected.")
```

**Task 3: Knowledge Base Creation (Rule Writing)**

```python
print("Rule-Based Expert System")

engine = input("Does the engine won't start? (yes/no): ")
battery = input("Is the battery low? (yes/no): ")

headache = input("Do you have a headache? (yes/no): ")
fever = input("Do you have fever? (yes/no): ")

internet = input("Is the internet slow? (yes/no): ")
ping = input("Is ping high? (yes/no): ")

temperature = input("Is the temperature high? (yes/no): ")
humidity = input("Is humidity high? (yes/no): ")

savings = input("Are your savings low? (yes/no): ")
salary = input("Is your salary low? (yes/no): ")

print("\n--- Expert System Result ---")

if engine == "yes" and battery == "yes":
    print("Diagnosis: Dead Battery")
elif headache == "yes" and fever == "yes":
    print("Diagnosis: Possible Infection")
elif internet == "yes" and ping == "yes":
    print("Diagnosis: Network Latency Issue")
elif temperature == "yes" and humidity == "yes":
    print("Diagnosis: Heat Index is High")
elif savings == "yes" and salary == "yes":
    print("Diagnosis: Loan Not Recommended")
else:
    print("No matching condition found.")
```
